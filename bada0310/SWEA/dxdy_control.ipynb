```python
# [SWEA] 16244
# 지정 box 의 테두리만 1로 출력하기 
T = int(input())
for t in range(1, T+1):
    grid = [[0]*10 for _ in range(10)]
    R, C, N = map(int,input().split())
    for i in range(N):
        for j in range(N):
            curr_r = R + i
            curr_c = C + j 

            if i == 0 or i == N-1 or j == 0 or j == N-1:
                grid[curr_r][curr_c] = 1
            else:
                grid[curr_r][curr_c] = 0
    print(f'#{t}')
    for line in grid:
        print(*line)
```

```python
# [SWEA] 16242
# 배열순회_ 지그재그 
T = int(input())
for t in range(1, T+1):
    grid = [[0]*10 for _ in range(10)]
    R, C, N, M = map(int,input().split())
    idx = 1
    for i in range(N):
        curr_r = R + i 
        if i%2 == 0:
            for j in range(M):
                grid[curr_r][C + j] = idx
                idx += 1
        else:
            for j in range(M-1, -1, -1):
                grid[curr_r][C+j] = idx
                idx += 1
    print(f'#{t}')
    for line in grid:
        print(*line)
```

```python
# [SWEA] 16240
# 배열순회_대각
T = int(input())
for t in range(1, T+1):
    grid = [[0]*10 for _ in range(10)]
    R, C, N = map(int,input().split())
    for r in range(R, R+N):
        grid[r][C + (r-R)] = 1
        grid[r][(C+N)-(r-R)-1] = 1
    print(f'#{t}')
    for line in grid:
        print(*line)
```
