```python
# bubble sort / counting sort  >> 안정형
# selection sort >> 불안정형 
# 안정형 정렬: 값이 같은 원소들이 정렬 후에도 입력 배열에서의 상대적 위치를 유지하는 정렬
# 불안정형 정렬: 값이 같은 원소들이 정렬 후에는 입력 배열에서의 상대적 위치가 유지되지 않는 경우 

# # bubble sort
# 시간 복잡도 = O(N^2)
# 공간 복잡도 = O(1)  # arr[i]~ arr[i+1]
# 효울적인 정렬 방법은 아님

# # selection sort 
# 시간 복잡도 = O(N^2) # T(n) = (n-1)+(n-2)+(n-3)+...+1 = (n-1)*n/2
# 공간 복잡도 = O(1)
# 기존 배열 이외의 추가적인 메모리를 거의 사용하지 않는 제자리 정렬(in-place sort)

# counting sort
# 시간 복잡도  = O(N+k)
# 공간 복잡도 = O(k) #len(lst)
# 데이터 범위가 작을 때. 유리 
```

```python
# Bubble sort
# 인접한 두 개의 원소를 비교하며 자리를 계속 교환하는 방식 
arr = [55, 7, 78, 12, 42]
def bubble_sort(arr, N): 
    for i in range(N-1,0,-1): # 범위의 끝 위치 
        for j in range(i): 
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j] # 두 인덱스의 값을 재정의 
    return arr

arr = [1, 4, 25, 213, 211, 1, 23, 13, 4, 5 ,1]

print(bubble_sort(arr,len(arr)))
```

```python
# selection sort
# function 
def selection_sort(arr,N):
    for i in range(N-1):
        min_idx = i
        for j in range(i+1, N):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]  # 두 인덱스의 값을 재정의
    # arr[i] 를 제외한 구간에서 탐색한 min_val = arr[j]를 
    # 기준인 arr[i] 의 위치와 arr[j] 의 위치를 바꾼다 .
    return arr
    
arr = [1, 4, 25, 213, 211, 1, 23, 13, 4, 5 ,1]

print(selection_sort(arr,10))
```

```python
#4차 수정 
#day03_실전문제 
# ROG 마을
from collections import deque
import sys
input = sys.stdin.readline

N, A = map(int,input().split())
board = [list(map(int,input().split())) for _ in range(N)]
x1, y1 = map(int,input().split()) # 코디의 초기 위치 
x2, y2 = map(int,input().split()) # 목적지 
visited = [[False]*N for _ in range(N)]
# R = 0 O = 1 G = 2

def bfs():
    queue = deque([x1,y1,0]) #시작점의 좌표 + time 
    
    visited = set([x1, y1,0]) # 방문 상태 
    
    while queue:
        curr_x, curr_y, time = queue.popleft()
        
        if curr_x == x2 and curr_y == y2:
            return time
        
        next_time = time + 1 #이동시 count 
        # 그리고 이동하지 않더라도 count (대기상태)
        
        change_count = next_time//A
        
        for dx, dy in [(0,1), (1,0),(0,-1),(-1,0)]:
            nx, ny =curr_x + dx, curr_y + dy
            
            if 0<= nx < N and 0<= ny < N:
                ori_color = board[nx][ny]
                curr_color = (ori_color + change_count)%3 # R = 0 O = 1 G = 2
                
            if curr_color == 2 and (nx, ny, next_time%(A*3)) not in visited:
                visited.add((nx, ny, next_time%(A*3)))
                queue.append((nx,ny, next_time))
    return -1
            
result = bfs()
print(result)
```
